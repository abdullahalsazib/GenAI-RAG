```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings = OllamaEmbeddings(
    base_url="http://localhost:11434",
    model="nomic-embed-text"
)
embeddings
```

```python
r1 = embeddings.embed_documents([
    "alpha is the first latter of grek alplhabe",
    "beta is the second latter of grek alplhabe",
])
r1
```

```python
embeddings.embed_query("what is the first latter of grek? ")
```
